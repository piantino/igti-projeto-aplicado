```python
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Q, Search

client = Elasticsearch()

search = Search(using=client, index='pa')
```

```python
def buscar_por_frase(search, frase):
    return search.query(Q({"match_phrase": {"ementa": frase}}))
```

```python
from datetime import datetime

def criar_analises(response):
    data = []
      
    for h in response:
        analise = {
            'numero': h.processo,
            'agente': 'script',
            'rotulo': 'SCO',
            'realizado': datetime.now().strftime('%Y-%m-%dT%H:%M:%S.%fZ')
        }
        data.append(analise)

    return data
```

```python
def executar_analises(search, cats, catss):

    for i in range(len(cats)):
        for frase in cats[i]:
            result = buscar_por_frase(search, frase)
            print("{} - {} - {}".format(catss[i], result.count(), frase))

            response = result.execute()
            data = criar_analises(response)
```
